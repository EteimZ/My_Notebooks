# Derivatives

Derivatives is a means of measuring the rate of change of a curve. [Slopes](https://en.wikipedia.org/wiki/Slope) in mathematics are a means of measuring the average rate of change. While derivatives measure the rate of change at an instance.

```python
import matplotlib.pyplot as plt
import numpy as np

m = 1 # slope

x = np.linspace(0, 50) # x component

y = m * x # y component

plt.title("The gradient of a line")
plt.grid()
plt.plot(x,y)
plt.plot([30, 10], [10,10], label="run")
plt.plot([30, 30], [30,10], label="rise")
plt.text(32, 30, '$(x_2, y_2)$', fontsize = 12)
plt.text(5, 7, '$(x_1, y_1)$', fontsize = 12)
plt.text(35, 20, '$\Delta y = y_2 - y_1$', fontsize = 12)
plt.text(15, 3, '$\Delta x = x_2 - x_1$', fontsize = 12)
plt.plot([30], [30], marker="o", markersize=8, markeredgecolor="red", markerfacecolor="green")
plt.plot([10], [10], marker="o", markersize=8, markeredgecolor="red", markerfacecolor="green")
plt.legend()
```

The slope also known as the gradient of a line is given by the equation:

$$\text{slope} = \frac{\text{rise}}{\text{run}}$$

$$\text{rise} = y_2 - y_1 $$

$$\text{run} = x_2 - x_1 $$

$$\text{slope} = \frac{y_2 - y_1}{x_2 - x_1}$$

```python
x = np.linspace(0, 5) # x component

y = x * x # y component


plt.title("The average slope of a curve")
plt.grid()
plt.plot(x,y)
plt.plot([2, 4], [4,16], label="avg slope")
plt.text(4.2, 16, '$(x_2, y_2)$', fontsize = 12)
plt.text(2.2, 3.8, '$(x_1, y_1)$', fontsize = 12)
plt.plot([2], [4], marker="o", markersize=8, markeredgecolor="red", markerfacecolor="green")
plt.plot([4], [16], marker="o", markersize=8, markeredgecolor="red", markerfacecolor="green")
plt.legend()
```

The average slope of a curve between two points can be calculated just like the gradient of a line

```python
import plotly.express as px
fig = px.line(x=x, y=y, markers=True, title='Derivative of a curve')
fig.show()
```

In other to calculate the slope of a point we need to use derivatives.

The formula for calculating derivative:

$$\frac{Δy}{Δx} = \frac{f(x + Δx) + f(x)}{Δx}$$

where:


$$ \textbf{x} \text{ is the range of values } $$
$$ \textbf{f(x) } \text{is a function applied to x} $$
$$ Δx \text{ is a tiny change in x} $$

```python
def sq(x): # function
  return x * x

h = 0.00001 # tiny change in x

x = np.linspace(0, 5) # input values

y = sq(x) # apply the function to x
der_y = (sq(x + h) - sq(x))/h # calculate the derivatives using the formula

plt.title("Derivative of the function $f(x) = x^2 $")
plt.grid()
plt.plot(x, y, label="function $x^2$")
plt.plot(x,der_y, label="derative of $x^2$")
plt.legend()
plt.show()
```
